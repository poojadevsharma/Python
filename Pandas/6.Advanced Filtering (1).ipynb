### Suppose you have a manager. If Profit is greater than 1000, Target Column is Achieved else Target Column Not Achieved

```python
df['target'] = 'Achieved'
df
```

```python
df.loc[df['Profit']<=500,'target'] = 'Not Achieved'
df
```
