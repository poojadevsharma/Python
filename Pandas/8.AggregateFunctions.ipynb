# Group By

```python
df = pd.read_excel('sample.xlsx')
df
```

```python
df.groupby('class').agg(
    avg_score = ('math','mean')
)
```

```python
df.groupby('class').agg(
    math_avg_score = ('math','mean'),
    phys_avg_score = ('phys','mean'),
    chem_avg_score = ('chem','mean'),
    math_max_score = ('math','max'),
    
    
    )
```
