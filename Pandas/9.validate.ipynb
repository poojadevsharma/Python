# 12 Numbers 

```python
df['ad copy'] = df['ad']
df['ad copy'] = df['ad copy'].str.replace('-','')
df['numbers'] = df['ad copy'].str.isnumeric()
df['length'] = df['ad copy'].str.len()
df = df.loc[ (df['numbers']==True)  &  (df['length']==12)  ]
df = df.drop(columns=['ad copy','numbers','length'])
df
```

```python
df = pd.read_excel('file.xlsx')
df['cid copy'] = df['cid']
df['cid copy'] = df['cid copy'].str.lower()
df['start'] = df['cid copy'].str.startswith('tcs')
df
```

```python
def f(v):
    if(v.startswith('tcs') and v[3]=='-' and len(v[4:])==2 and v[4:].isnumeric()):
        return True 
    else:
        return False

df['new'] = df['cid copy'].apply(f)   
df

```
