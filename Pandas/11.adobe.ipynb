# Calculate Department Wise Annual Savings 

```python
import pandas as pd
```

```python
budget = pd.read_excel('adobe.xlsx',sheet_name='Budget')
budget
```

```python
expenditure = pd.read_excel('adobe.xlsx',sheet_name='Expenditure')
expenditure
```

```python
budget['Department'] = budget['Department'].str.title()  
budget
```

```python
expenditure.columns = expenditure.columns.str.strip()
expenditure.columns
```

```python
expenditure['Department'] = expenditure['Department'].str.title()
expenditure
```

```python
df = budget.merge(expenditure,on=['Date','Department'],how='inner')
df['Savings'] = df['Budget'] - df['Actual Spend']
df
```

```python
df['Department'] = df['Department'].str.strip()
df['Department'].unique()
```

```python
ndf = df.groupby('Department').agg(
    annual_saving = ('Savings','sum')
)
ndf
```
