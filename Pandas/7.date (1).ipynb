```python
import pandas as pd
```

```python
df = pd.read_csv('Walmart_Sales.csv')
df['Date'] = df['Date'].str.replace('/','-')
df
```

```python
df.info()
```

# Conversion to date time - Year , month , day

```python
df['Date'] = pd.to_datetime(df['Date'],format='%d-%m-%Y')
df
```

Simple Functions

```python
df['Date'].min()
```

```python
df['Date'].max()
```

```python
df = df.sort_values('Date')
df
```

# Filtering

```python
ndf = df.loc[ df['Date']>'2010-02-05' ]
ndf
```

```python
ndf = df.loc[ df['Date']=='2010-02-05' ]
ndf
```

```python
ndf = df.loc[ (df['Date']>='2010-01-01')  &  (df['Date']<='2010-12-31') ]
ndf
```

# dt 

find day

```python
df['Day'] = df['Date'].dt.day
df
```

# month

```python
df['Month'] = df['Date'].dt.month  
df
```

# year

```python
df['Year'] = df['Date'].dt.year   
df
```

day name

```python
df['Day Name'] = df['Date'].dt.day_name()   
df
```

```python
df['Month Name'] = df['Date'].dt.month_name()   
df
```

dayofweek

```python
df['dow'] = df['Date'].dt.day_of_week 
df
```

```python
df['dim'] = df['Date'].dt.daysinmonth 
df
```
