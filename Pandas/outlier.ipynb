```python
df=pd.read_excel('amazon.xlsx')
df
```

```python
df['Sales'].mean()
```

```python
df['Sales'].median()
```

```python
q1,q3=np.percentile(df['Sales'],[25,75])
iqr=q3-q1
lx=q1-1.5*iqr
ux=q3+1.5*iqr
df.loc[df['Sales']>ux]
```

```python
outliers=df.loc[(df['Sales']<lx)|(df['Sales']>ux)]
outliers
```

```python
ndf=df.loc[(df['Sales']>lx)&(df['Sales']<ux)]
ndf
```

```python
ndf['Sales'].mean()
```

```python
ndf['Sales'].median()
```
