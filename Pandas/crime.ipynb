```python
ndf=df.loc[df['Vict Age']>0]
ndf
```

```python
def f(v):
    if v>0 and v<10:
        return '0-10'
    elif v>10 and v<=20:
        return '10_20'
    elif v>20 and v<=30:
        return '20_30'
    elif v>30 and v<=40:
        return '30_40'
    elif v>40 and v<=50:
        return '40_50'
    elif v>50 and v<=60:
        return '50_60'
    elif v>60 and v<=70:
        return '60_70'
    elif v>70 and v<=80:
        return '70_80'
    elif v>80 and v<=90:
        return '80_90'
    elif v>90 and v<=100:
        return '90_100'
    elif v>100 and v<=110:
        return '100_110'
    elif v>110 and v<=120:
        return '110_120'
    
ndf['Age']=ndf['Vict Age'].apply(f)
ndf

 
```

```python
ndf['Age'].value_counts().reset_index()
```

```python
ndf=ndf.groupby('Age').agg(
    crime=('Crm Cd Desc', 'sum')
)
ndf
```
