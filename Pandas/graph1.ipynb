```python
students=['pawan','shivam','shubham','pooja','abhi']
math_scores=[20,30,45,57,48]
phy_score=[30,40,50,10,20]
biology_score=[30,20,31,34,35]
plt.figure(figsize=(10, 6))
plt.scatter(students,math_scores,color='red',s=600,alpha=1,label='maths')
plt.scatter(students,phy_score,color='blue',s=600,label='physics')
plt.scatter(students,biology_score,color='green',s=600,label='biology')
plt.title('student_score',color='green',fontsize=15)
plt.xlabel('(students)')
plt.ylabel('(math_scores)',color='blue')
plt.xticks(rotation=25,color='purple',fontsize=10)
plt.yticks(color='purple')
plt.grid()
for m,n in zip(students,math_scores):
    plt.text(x=m,y=n,s=n,ha='center',va='top',color='black',fontsize=13)
for m,n in zip(students,phy_score):
    plt.text(x=m,y=n,s=n,ha='center',va='top',color='black',fontsize=13)
for m,n in zip(students,biology_score):
    plt.text(x=m,y=n,s=n,ha='center',va='top',color='black',fontsize=13)
plt.legend(bbox_to_anchor=(1,1),loc='upper left')
plt.show()
```

```python
month=['jan','feb','mar','apr','may','jun']
isales=[200,300,450,500,600,700]
ssales=[300,400,200,450,650,800]

plt.figure(figsize=(10,5))
plt.plot(month,isales,color='green',marker='o',markerfacecolor='red',markeredgecolor='black',markersize=8,label='isales')
plt.plot(month,ssales,color='magenta',marker='o',markerfacecolor='blue',markeredgecolor='black',markersize=8,label='ssales')
plt.grid(color='red')
plt.title('monthly sales')
plt.xlabel('month')
plt.ylabel('sales')
plt.legend()
for m,n in zip(month,isales):
    plt.text(x=m,y=n,s=n,ha='center',va='bottom',color='black',fontsize=13)
for m,n in zip(month,ssales):
    plt.text(x=m,y=n,s=n,ha='center',va='bottom',color='black',fontsize=13)



plt.show()
```

```python
import plotly.graph_objects as go

month = ['jan', 'feb', 'mar', 'apr', 'may', 'jun']
isales = [200, 300, 450, 500, 600, 700]
ssales = [300, 400, 200, 450, 650, 800]

fig = go.Figure()

# Add isales trace
fig.add_trace(go.Scatter(
    x=month,
    y=isales,
    mode='lines+markers+text',
    name='isales',
    line=dict(color='green'),
    marker=dict(
        color='red',
        size=12,
        line=dict(
            color='black',
            width=2
        )
    ),
    text=isales,
    textposition='top center',
    textfont=dict(size=13)
))

# Add ssales trace
fig.add_trace(go.Scatter(
    x=month,
    y=ssales,
    mode='lines+markers+text',
    name='ssales',
    line=dict(color='magenta'),
    marker=dict(
        color='blue',
        size=12,
        line=dict(
            color='black',
            width=2
        )
    ),
    text=ssales,
    textposition='top center',
    textfont=dict(size=13)
))

# Update layout
fig.update_layout(
    title='Monthly Sales',
    xaxis_title='Month',
    yaxis_title='Sales',
    showlegend=True,
    width=1000,
    height=500,
    plot_bgcolor='white'
)

# Add grid
fig.update_xaxes(showgrid=True, gridwidth=1, gridcolor='red')
fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='red')

fig.show()
```

```python
import plotly.graph_objects as go

ing = ['apple', 'banana', 'orange', 'grapes', 'mango']
quantity = [20, 30, 15, 25, 35]

fig = go.Figure(data=[go.Pie(labels=ing, values=quantity, textinfo='label+percent')])

fig.update_traces(textposition='inside', textinfo='percent+label')
fig.update_layout(title='Fruit Quantities')

fig.show()
```
